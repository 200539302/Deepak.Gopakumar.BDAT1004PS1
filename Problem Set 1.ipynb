# Question 3

#### Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
#### sides of a triangle and returns the area of the triangle. By Heron's formula, the area 
#### of a triangle with side lengths a, b, and c is
#### s(s - a)(s -b)(s -c)
#### , where 
#### s = (a+b+c)/2. 


```python
using System;

namespace ConsoleApp1
{
    // Used the below link to find how to convert string to double while accepting user input
    // https://stackoverflow.com/questions/11399439/converting-string-to-double-in-c-sharp
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine("Please enter the dimensions of the triangle to calculate the area:");
            Console.Write("Please enter side A: ");
            double sideA = Convert.ToDouble(Console.ReadLine());

            Console.Write("Please enter side B: ");
            double sideB = Convert.ToDouble(Console.ReadLine());

            Console.Write("Please enter side C: ");
            double sideC = Convert.ToDouble(Console.ReadLine());

            double area = triangleArea(sideA, sideB, sideC);

            Console.WriteLine("Area of the triangle with the given dimensions is: {0}", area);
            Console.ReadLine();
        }

        public static double triangleArea(double a, double b, double c)
        {
            double semiperimeter = (a + b + c) / 2;
            double area = Math.Sqrt(semiperimeter * (semiperimeter - a) * (semiperimeter - b) * (semiperimeter - c));
            return area;
        }
    }
}
```

# Question 4
#### Write a program in C# Sharp to separate odd and even integers in separate arrays. 


```python
using System;

class Program
{
    static void Main(string[] args)
    {
        Console.Write("Enter the total number of elements in the array: ");
        int totalCount = Convert.ToInt32(Console.ReadLine());

        Console.Write("Enter {0} elements in the array:\n", totalCount);
        int[] elements = new int[totalCount];

        for (int i = 0; i < totalCount; i++)
        {
            Console.Write("Element - {0}: ", i);
            elements[i] = Convert.ToInt32(Console.ReadLine());
        }

        int[] evenNumbers = GetEvenElements(elements);
        int[] oddNumbers = GetOddElements(elements);

        Console.WriteLine("The Even elements are:");
        DisplayArray(evenNumbers);

        Console.WriteLine("The Odd elements are:");
        DisplayArray(oddNumbers);

    }

    public static int[] GetEvenElements(int[] numbers)
    {
        int count = CountEvenElements(numbers);
        int[] evenArray = new int[count];
        int index = 0;

        for (int i = 0; i < numbers.Length; i++)
        {
            if (IsEven(numbers[i]))
            {
                evenArray[index] = numbers[i];
                index++;
            }
        }

        return evenArray;
    }

    public static int[] GetOddElements(int[] numbers)
    {
        int count = CountOddElements(numbers);
        int[] oddArray = new int[count];
        int index = 0;

        for (int i = 0; i < numbers.Length; i++)
        {
            if (IsOdd(numbers[i]))
            {
                oddArray[index] = numbers[i];
                index++;
            }
        }

        return oddArray;
    }

    public static int CountEvenElements(int[] numbers)
    {
        int count = 0;

        foreach (int num in numbers)
        {
            if (IsEven(num))
            {
                count++;
            }
        }

        return count;
    }

    public static int CountOddElements(int[] numbers)
    {
        int count = 0;

        foreach (int num in numbers)
        {
            if (IsOdd(num))
            {
                count++;
            }
        }

        return count;
    }

    public static bool IsEven(int number)
    {
        return number % 2 == 0;
    }

    public static bool IsOdd(int number)
    {
        return number % 2 != 0;
    }

    public static void DisplayArray(int[] array)
    {
        foreach (int number in array)
        {
            Console.Write(number + " ");
        }
        Console.WriteLine();
    }
}
```

#### Write a pig Latin function


```python
def pig(word):
    # Convert the word to lowercase
    word = word.lower()

    # Check if the word starts with a vowel
    vowels = ['a', 'e', 'i', 'o', 'u']
    if word[0] in vowels:
        return word + 'way'
    else:
        return word[1:] + word[0] + 'ay'


def main():
    # Prompt the user for input
    word = input("Please enter a word: ")

    # Call the pig() function to convert the word to pig Latin
    pig_word = pig(word)
    print(f"Pig Latin translation: {pig_word}")


if __name__ == '__main__':
    main()
```

```python
from collections import Counter

def bldcount(file_path):
   
        with open(file_path, 'r') as input_file:
            blood_types = input_file.read().split()
            counts = Counter(blood_types)
            blood_type_mapping = {
                'A': 'blood type A',
                'B': 'blood type B',
                'AB': 'blood type AB',
                'O': 'blood type O',
                'OO': 'blood type OO'
            }

        for blood_type, count in counts.items():
            if blood_type in blood_type_mapping:
                blood_type_description = blood_type_mapping[blood_type]
                if count == 1:
                    print(f"There is one patient with {blood_type_description}.")
                else:
                    print(f"There are {count} patients with {blood_type_description}.")
            else:
                print(f"There are no patients with blood type {blood_type}.")


bldcount('bloodtype1.txt')
```

# Question 8

#### Write a function curconv()


```python
def curconv(target_currency, amount):
    currency_data = load_currency_data()
    exchange_rate = find_exchange_rate(currency_data, target_currency)

    if exchange_rate is not None:
        usd_amount = calculate_usd_equivalent(amount, exchange_rate)
        return usd_amount

    return "Currency not found in the data."


def load_currency_data():
    with open("currencies.txt", "r") as file:
        currency_data = {}

        for line in file:
            data_list = line.split()
            currency = data_list[0]
            exchange_rate = float(data_list[1])
            currency_data[currency] = exchange_rate

    return currency_data


def find_exchange_rate(currency_data, target_currency):
    if target_currency in currency_data:
        return currency_data[target_currency]

    return None


def calculate_usd_equivalent(amount, exchange_rate):
    usd_amount = amount * exchange_rate
    return usd_amount


print(curconv("EUR", 100))
print(curconv("JPY", 100))
```

# Question 10 

#### Write a function called frequencies()

```python
def  frequencies(text):
    char_frequencies = [0] * 26
    for char in text:
        ascii_code = ord(char)
        if ascii_code >= ord('a') and ascii_code <= ord('z'):
            index = ascii_code - ord('a')
            char_frequencies[index] += 1
    return char_frequencies

if __name__ == '__main__':
    input_text1 = 'The quick red fox got bored and went home.'
    input_text2 = "apple"
    char_freqs1 =  frequencies(input_text1)
    char_freqs2 =  frequencies(input_text2)

    print(char_freqs1)
    print(char_freqs2)
```
